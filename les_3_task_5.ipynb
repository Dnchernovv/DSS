```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
```

```python
data = load_wine()
data_keys = data.keys()
print(type(data))
```

```python
print(len(data['target_names']), data['feature_names'], data['DESCR'])
```

```python
X = pd.DataFrame(data = data['data'], columns = data['feature_names'])
print(X.shape)
X.isnull().any()
```

```python
X['target'] = data['target']
```

```python
X['target'] = data['target']
X_corr = X.corr()

high_corr = X_corr.loc['target']
high_corr = high_corr[(abs(high_corr) > 0.5) & (high_corr != 1.0)]
high_corr = high_corr.index
```

```python
X.drop(columns = 'target', inplace = True)
```

```python
suff = '_2'
suffixed = X.columns[X.columns.str.contains('_2')]
for x in high_corr:
    X.loc[X[x] == X[x], x + suff] = 0
for i in range(len(high_corr)):
    X.loc[X[suffixed[i]] == X[suffixed[i]], suffixed[i]] = (X[high_corr[i]]) ** 2
```

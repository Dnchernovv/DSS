```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
plt.style.use('fivethirtyeight')
```

```python
cr = pd.read_csv('creditcard.csv')
cr
```

```python
fig, ax = plt.subplots(ncols = 2, nrows = 1)
ax1, ax2 = ax.flatten()
fig.subplots_adjust(hspace = 0.7, wspace = 0.7)
ax1.bar(cr['Class'].unique(),cr['Class'].value_counts())
ax2.bar(cr['Class'].unique(),cr['Class'].value_counts(), log = True);
```

```python
cr.loc[cr['Class'] == 0,'V1']
cr.loc[cr['Class'] == 1,'V1']
```

```python
fig, ax = plt.subplots(ncols = 2, nrows = 1)
ax1, ax2 = ax.flatten()

fig.subplots_adjust(hspace = 0.7, wspace = 0.7)

ax1.hist(cr.loc[cr['Class'] == 0,'V1'], density = True, bins = 20, alpha = 0.5, color = 'grey', label = 'Class 0')
ax1.set_xlabel('Class')
ax1.legend(loc = 'best')

ax2.hist(cr.loc[cr['Class'] == 1, 'V1'], density = True, bins = 20, alpha = 0.5, color = 'red', label = 'Class 1')

ax2.set_xlabel('Class')
ax2.legend(loc="best");
```
